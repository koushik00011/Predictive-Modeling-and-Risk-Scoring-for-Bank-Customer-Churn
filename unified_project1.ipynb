```python
X = df.drop("exited", axis=1)

y = df["exited"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
model = LogisticRegression(
    class_weight="balanced"
)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    random_state=42
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
})

feature_importance = feature_importance.sort_values(
    by="Importance",
    ascending=False
)

feature_importance.head(10)
```

```python
#1.senior customers had highest churn rates so the age stringly influence the churn behaviour its the same as sql analysisso the accuracy is good.
#2.credit_score also influence the churn behaviour so financial profile affects retention probability.
#3.Income level influence the churn behaviour due to different financial expectations or service dissatisfaction.
#4.medium/high-balance customers had high churn
#5.4-product customers have high churn rate
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_prob = rf_model.predict_proba(X_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
```

```python
roc_auc = auc(fpr, tpr)

print(roc_auc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.plot(
    fpr,
    tpr,
    label=f"AUC = {roc_auc:.2f}"
)

plt.plot([0,1], [0,1], linestyle="--")
 
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title("ROC Curve - Random Forest")

plt.legend()

plt.show()
```

```python
#The ROC curve demonstrated strong classification capability with an AUC score of 0.85. The model achieved effective separation between churn and 
#non-churn customers, significantly outperforming random classification.
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(
    random_state=42
)

gb_model.fit(X_train, y_train)

y_pred_gb = gb_model.predict(X_test)

print(classification_report(y_test, y_pred_gb))
```
